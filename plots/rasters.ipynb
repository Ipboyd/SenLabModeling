```python
import matplotlib.pyplot as plt
```

```python
import pickle as pk

file_path = '../py_spike_data.pkl'
with open(file_path, 'rb') as file:
    spks = pk.load(file)
```

```python
keys = list(spks.keys())
spks_1 = spks[keys[0]]
wide_sequence = spks_1['stimulus_0_poisson_spks'][:,0,:].transpose()
```

```python
fig, ax = plt.subplots(figsize=(12, 3))
ax.spy(wide_sequence, markersize=.5)
ax.set_aspect('auto')
ax.set_title("Rasters")
ax.set_xlabel("Time step")
ax.set_ylabel("Trial")

plt.tight_layout()
plt.show()
```
